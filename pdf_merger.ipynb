```python
import PyPDF2
```

```python
from PyPDF2 import PdfMerger

# Create an instance of PdfMerger
merger = PdfMerger()
pdfs = []
n = int(input("How many pdfs do you want to merge?\n"))

for i in range(n):
    # Ask for full path or ensure file is in current directory
    name = input(f"Enter the full path or name of pdf{i+1} (include .pdf extension): ")
    
    # Check if file exists before adding to the list
    import os
    if os.path.exists(name):
        pdfs.append(name)
    else:
        print(f"Warning: File '{name}' not found. Skipping this file.")

# Only proceed if we have PDFs to merge
if pdfs:
    for pdf in pdfs:
        merger.append(pdf)
    
    merger.write("merged-pdf.pdf")
    merger.close()
    print("PDFs merged successfully!")
else:
    print("No valid PDF files to merge.")
```
